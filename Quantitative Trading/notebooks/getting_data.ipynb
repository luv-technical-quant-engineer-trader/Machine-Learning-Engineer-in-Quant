##### Getting data from Yahoo Finance

- Using YFinance library to get data which internally use Yahoo Finance library
- Download is used when we want to get multiple ticker's data.

```python
data = yf.Ticker("ARCC")
arguments_list = ["info","calendar","analyst_price_targets", "quarterly_income_stmt"] #We can use option chain and history as well
```

```python
for arguments in arguments_list:

        value = getattr(data, arguments)
        print(f"--- {arguments.upper()} ---")
        print(value)
        print("\n")
```
